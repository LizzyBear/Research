```python
# import 各种库
import requests
from bs4 import BeautifulSoup
import csv
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
def get_journals(url):
    global journals
    
    response = requests.get(url)
    html_soup = BeautifulSoup(response.text,'html.parser')
    journals = html_soup.find_all('div', class_ ="nova-v-publication-item__stack nova-v-publication-item__stack--gutter-m")
    
    return journals
```

```python
title_list = []
```

```python
date_list = []
```

```python
abstract_list = []
```

```python
def analyze_journal(journal,title_list,date_list,abstract_list):
    
    global title,date,abstract
    # get title
    title = journal.find('a', class_="nova-e-link nova-e-link--color-inherit nova-e-link--theme-bare").get_text()
    # get date
    date = journal.find('li',class_="nova-e-list__item publication-item-meta-items__meta-data-item").find('span').get_text()
    # get abstract
    abs = journal.find('div',class_="nova-e-text nova-e-text--size-m nova-e-text--family-sans-serif nova-e-text--spacing-none nova-e-text--color-inherit nova-v-publication-item__description nova-v-publication-item__description--clamp-3")
    if abs == None:
        abstract = 'no abstract provided'
    else:
        abstract = abs.get_text()
    
    title_list.append(title)
    date_list.append(date)
    abstract_list.append(abstract)
    
    return title_list,date_list,abstract_list
```
